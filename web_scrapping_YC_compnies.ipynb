```python
# Extracting company name and locations
company_name=[]
company_location=[]
driver.implicitly_wait(25)
for i in range(1,250):
    try:
        company_name_element= driver.find_element('xpath','/html/body/div/div[2]/section[2]/div/div[2]/div[2]/div[4]/a[{0}]/div/div[2]/div/div[1]/span[1]'.format(i))
        company_name.append(company_name_element.text)
    except:
        company_name.append("NAN")
    
    try:
        company_location_element= driver.find_element('xpath','/html/body/div/div[2]/section[2]/div/div[2]/div[2]/div[4]/a[{0}]/div/div[2]/div/div[1]/span[2]'.format(i))
        company_location.append(company_location_element.text)
    except:
        company_location.append("NAN")
        

    
```

```python
import pandas as pd
```

```python
# Making the dataframe from company name and Comapny location
data = {'Company Name': company_name, 'Company Location': company_location}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
# replacing '.' to '-' from the company name because it will not coming in the company page link, so to extract info
# from there we have to make links and it will be required there.
# similarly remove '()', make all latters small, and fill space by '-' for generating the links
df['Company Name'] = df['Company Name'].str.replace(r'\b(\w+)\.(\w+)\b', r'\1-\2', regex=True)
```

```python
# saving company name list and locations to a csv file
df.to_csv('Companies_list.csv', index= False)
```

```python
# Extracting all other info which are asked in 2nd point of the assignment.
# like Comapny name, location, founded in, team size, founders 
Company_list=[]
Founded_in=[] 
Team_size=[]
Location=[]
Group_partner=[]
Link=[]
Active_founder=[]
# driver.implicitly_wait(10)
base_url="https://www.ycombinator.com/companies/{}"
for index, row in df.iterrows():
    # Get the page number from the DataFrame column
    company= row['Company Name']
    
    # Create the dynamic URL for each row
    url = base_url.format(company)
#     print(url)
    
    driver.get(url)
    # Company name
    try:      
        company_name_element= driver.find_element('xpath','/html/body/div/div[2]/section[1]/div[2]/div[1]/div[1]/div[2]/div[1]/h1')
        Company_list.append(company_name_element.text)
    except:
        Company_list.append("NaN")
    # Company founded year
    try:
        company_founded_element= driver.find_element('xpath','/html/body/div/div[2]/section[1]/div[2]/div[2]/div/div[2]/div[1]/span[2]')
        Founded_in.append(company_founded_element.text)
    except:
        Founded_in.append('NaN')
    # Team size of the company    
    try:
        company_team_size_element= driver.find_element('xpath','/html/body/div/div[2]/section[1]/div[2]/div[2]/div/div[2]/div[2]/span[2]')
        Team_size.append(company_team_size_element.text)
    except:
        Team_size.append('NaN')
        
    # Companie's location
    try:
        company_location_element= driver.find_element('xpath','/html/body/div/div[2]/section[1]/div[2]/div[2]/div/div[2]/div[3]/span[2]')
        Location.append(company_location_element.text)
    except:
        Location.append('NaN')
        
    # Group Partner
    try:
        company_group_partner_element= driver.find_element('xpath','/html/body/div/div[2]/section[1]/div[2]/div[2]/div/div[2]/div[4]/a')
        Group_partner.append(company_group_partner_element.text)
    except:
        Group_partner.append('NaN')
        
    # Company link
    try:
        company_link_element= driver.find_element('xpath','/html/body/div/div[2]/section[1]/div[2]/div[1]/div[2]/div/div/a/div')
        Link.append(company_link_element.text)
    except:
        Link.append('NaN')
        
    # Active founders
    for i in range(1,3):
        try:
            active_founders_element= driver.find_element('xpath','/html/body/div/div[2]/div/section/div[2]/div[{0}]/div[1]/h3'.format(i))
            Active_founder.append(active_founders_element.text)
        except:
            Active_founder.append('NaN')  
```
